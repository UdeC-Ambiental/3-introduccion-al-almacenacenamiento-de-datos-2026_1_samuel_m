### 💡 Actividad
- Crea al menos dos condiciones (por pH y temperatura).
- Genera una nueva columna con la categoría resultante.

```python
# import pandas as pd

datos = pd.read_csv('../data/4_resultados_agua.csv')

print("Columnas:", datos.columns)

# pH fuera de rango
ph_fuera = datos[(datos['pH'] < 6.5) | (datos['pH'] > 8.5)]
print("\npH fuera de rango:")
print(ph_fuera)

# Temperatura alta (USANDO Temp)
temp_alta = datos[datos['Temp'] > 25]
print("\nTemperatura alta:")
print(temp_alta)

# Oxígeno bajo (USANDO OD_promedio)
ox_bajo = datos[datos['OD_promedio'] < 5]
print("\nOxígeno bajo:")
print(ox_bajo)

# Condiciones óptimas
optimo = datos[
    (datos['pH'] >= 6.5) &
    (datos['pH'] <= 8.5) &
    (datos['Temp'] < 25) &
    (datos['OD_promedio'] > 5)
]

print("\nCondiciones óptimas:")
print(optimo)
```
