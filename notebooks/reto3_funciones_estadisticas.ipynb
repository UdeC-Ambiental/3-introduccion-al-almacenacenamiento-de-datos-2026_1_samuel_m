### 🔹 Ejemplo de función personalizada
Crea una función que clasifique el pH:

```python
def clasificar_ph(valor):
    if valor < 6.5:
        return 'Ácido'
    elif valor <= 8.5:
        return 'Neutro'
    else:
        return 'Básico'

datos['categoria_ph'] = datos['pH'].apply(clasificar_ph)
datos.head()
```

### 💡 Actividad
- Calcula promedios de variables numéricas.
- Aplica al menos una función con `apply()` o `groupby()`.

```python
# Tu código aquí import pandas as pd

# 1. Cargar datos
datos = pd.read_csv('../data/4_resultados_agua.csv')

print("Columnas:", datos.columns)

# 2. Estadísticas generales
print("\nEstadísticas descriptivas:")
print(datos.describe())

# 3. Promedios
print("\nPromedios:")
print("Promedio pH:", datos['pH'].mean())
print("Promedio Temperatura:", datos['Temp'].mean())
print("Promedio Oxígeno disuelto:", datos['OD_promedio'].mean())

# 4. Agrupación por sitio (si existe)
if 'Sitio' in datos.columns:
    print("\nPromedio por sitio:")
    print(datos.groupby('Sitio').mean(numeric_only=True))

# 5. Crear columna de calidad del agua
def clasificar_calidad(fila):
    if (fila['pH'] >= 6.5 and fila['pH'] <= 8.5 and
        fila['Temp'] < 25 and
        fila['OD_promedio'] > 5):
        return 'Buena'
    else:
        return 'Mala'

datos['Calidad_Agua'] = datos.apply(clasificar_calidad, axis=1)

# 6. Ver resultados
print("\nDatos con calidad del agua:")
print(datos.head())

# 7. Conteo de calidad
print("\nConteo de calidad:")
print(datos['Calidad_Agua'].value_counts())
```
