# 🧩 Reto 4 – Integración final de resultados

**Objetivo:** combinar los resultados de los retos anteriores para obtener un diagnóstico integral de la calidad del agua.

### 🔹 Instrucciones
- Importa los resultados obtenidos en los notebooks anteriores.
- Integra los análisis (pH, temperatura, oxígeno) en un solo DataFrame.
- Realiza una conclusión general del estado del agua.

```python
import pandas as pd

datos = pd.read_csv('../data/4_resultados_agua.csv')

print("Columnas:", datos.columns)
```

```python
# Filtros
ph_fuera = datos[(datos['pH'] < 6.5) | (datos['pH'] > 8.5)]
temp_alta = datos[datos['Temp'] > 25]
ox_bajo = datos[datos['OD_promedio'] < 5]

# Condiciones óptimas
optimo = datos[
    (datos['pH'] >= 6.5) &
    (datos['pH'] <= 8.5) &
    (datos['Temp'] < 25) &
    (datos['OD_promedio'] > 5)
]
```

```python
def clasificar_calidad(fila):
    if (fila['pH'] >= 6.5 and fila['pH'] <= 8.5 and
        fila['Temp'] < 25 and
        fila['OD_promedio'] > 5):
        return 'Buena'
    else:
        return 'Mala'

datos['Calidad_Agua'] = datos.apply(clasificar_calidad, axis=1)
```

```python
print("\nResumen de calidad del agua:")
print(datos['Calidad_Agua'].value_counts())
```

```python
if 'Sitio' in datos.columns:
    print("\nDiagnóstico por sitio:")
    print(datos.groupby('Sitio')['Calidad_Agua'].value_counts())
```

```python
print("\nEstadísticas generales:")
print(datos.describe())
```

A partir del análisis de los datos, se identificaron registros con condiciones no óptimas en pH, temperatura y oxígeno disuelto.

Se observó que una parte de los datos presenta pH fuera del rango permitido, lo cual puede afectar la vida acuática. Asimismo, temperaturas elevadas y bajos niveles de oxígeno disuelto indican posibles problemas de contaminación o alteración del ecosistema.

Mediante la clasificación de la calidad del agua, se determinó qué proporción de los registros se encuentra en condiciones adecuadas (“Buena”) y cuáles requieren atención (“Mala”).

Finalmente, el análisis por sitio permite identificar zonas críticas donde se deben priorizar acciones de monitoreo y mejora ambiental.
